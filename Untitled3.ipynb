```python
# Specify the columns for disease and outcome
disease_column = 'Disease'      # Replace 'Disease' with the actual disease column name
outcome_column = 'Outcome Variable'      # Replace 'Outcome' with the actual outcome variable column name

# Count the number of occurrences for each outcome category
outcome_counts = df[outcome_column].value_counts()

# Calculate percentages
outcome_percentage = (outcome_counts / outcome_counts.sum()) * 100

# Create a pie chart
labels = outcome_percentage.index.tolist()
sizes = outcome_percentage.values.tolist()
colors = ['lightblue', 'lightgreen', 'lightyellow', 'lightcoral']  # Add more colors if needed

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Outcome Variable Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Specify the columns for disease and cough
disease_column = 'Disease'  # Replace 'Disease' with the actual disease column name
cough_column = 'Cough'      # Replace 'Cough' with the actual cough column name

# Specify the disease of interest
target_disease = 'Asthma'

# Filter the data for people with Asthma
asthma_data = df[df[disease_column] == target_disease]

# Count the number of people with and without cough among those with Asthma
cough_counts = asthma_data[cough_column].value_counts()

# Calculate percentages
cough_percentage = (cough_counts / cough_counts.sum()) * 100

# Create a pie chart
labels = ['Cough', 'No Cough']
sizes = [cough_percentage.get('Yes', 0), cough_percentage.get('No', 0)]
colors = ['lightblue', 'lightgreen']

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title(f'Percentage of People with Asthma with and without Cough')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
